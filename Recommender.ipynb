### Building The Recommender

```python
# copy df
dfPredicted = dataR3T.head(1000)

# find the nearest neighbors using NearestNeighbors(n_neighbors=3)
number_neighbors = 3
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(dataR3T.values)
distances, indices = knn.kneighbors(dataR3T.values, n_neighbors=number_neighbors)

# convert user_name to user_index
user_index = dataR3T.columns.tolist().index(dataR3T.columns[5])

# t: game_title, m: the row number of t in dataR3T
for m,t in list(enumerate(dataR3T.index)):
  
  # find games without ratings by 53875128
  if dataR3T.iloc[m, user_index] == 0:
    sim_games = indices[m].tolist()
    games_distances = distances[m].tolist()
    
    # Generally, this is the case: indices[3] = [3 6 7]. The game itself is in the first place.
    # In this case, we take off 3 from the list. Then, indices[3] == [6 7] to have the nearest NEIGHBORS in the list. 
    if m in sim_games:
      id_game = sim_games.index(m)
      sim_games.remove(m)
      games_distances.pop(id_game) 

    # However, if the percentage of ratings in the dataset is very low, there are too many 0s in the dataset. 
    # Some games have all 0 ratings and the games with all 0s are considered the same games by NearestNeighbors(). 
    # Then,even the game itself cannot be included in the indices. 
    # For example, indices[3] = [2 4 7] is possible if game_2, game_3, game_4, and game_7 have all 0s for their ratings.
    # In that case, we take off the farthest game in the list. Therefore, 7 is taken off from the list, then indices[3] == [2 4].
    else:
      sim_games = sim_games[:number_neighbors-1]
      games_distances = games_distances[:number_neighbors-1]
        
    # game_similarty = 1 - game_distance    
    games_similarity = [1-x for x in games_distances]
    games_similarity_copy = games_similarity.copy()
    nominator = 0

    # for each similar game
    for s in range(0, len(games_similarity)):
      
      # check if the rating of a similar game is zero
      if dataR3T.iloc[sim_games[s], user_index] == 0:

        # if the rating is zero, ignore the rating and the similarity in calculating the predicted rating
        if len(games_similarity_copy) == (number_neighbors - 1):
          games_similarity_copy.pop(s)
          
        else:
          games_similarity_copy.pop(s-(len(games_similarity)-len(games_similarity_copy)))

      # if the rating is not zero, use the rating and similarity in the calculation
      else:
        nominator = nominator + games_similarity[s]*dataR3T.iloc[sim_games[s],user_index]

    # check if the number of the ratings with non-zero is positive
    if len(games_similarity_copy) > 0:
      
      # check if the sum of the ratings of the similar games is positive.
      if sum(games_similarity_copy) > 0:
        predicted_r = nominator/sum(games_similarity_copy)

      # Even if there are some games for which the ratings are positive, some games have zero similarity even though they are selected as similar games.
      # in this case, the predicted rating becomes zero as well  
      else:
        predicted_r = 0

    # if all the ratings of the similar games are zero, then predicted rating should be zero
    else:
      predicted_r = 0

  # place the predicted rating into the copy of the original dataset
    dfPredicted.iloc[m,user_index] = predicted_r

dfPredicted
```

```python
def recommend_games(user, num_recommended_games):

  print('The list of the games {} Has Played \n'.format(user))

  for m in dataR3T[dataR3T[user] > 0][user].index.tolist():
    print(m)
  
  print('\n')

  recommended_games = []

  for m in dataR3T[dataR3T[user] == 0].index.tolist():

    index_dataR3T = dataR3T.index.tolist().index(m)
    predicted_rating = dfPredicted.iloc[index_dataR3T, dfPredicted.columns.tolist().index(user)]
    recommended_games.append((m, predicted_rating))

  sorted_rm = sorted(recommended_games, key=lambda x:x[1], reverse=True)
  
  print('The list of the Recommended games \n')
  rank = 1
  for recommended_game in sorted_rm[:num_recommended_games]:
    
    print('{}: {} - predicted rating:{}'.format(rank, recommended_game[0], recommended_game[1]))
    rank = rank + 1
```

```python
# Get A Small List Of Recommended Games For One User
recommend_games(dataR3T.columns[0], 5)
```
